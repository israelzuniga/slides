### Control de flujo

#### If
"If" se utiliza para la ejecuci√≥n condicional, es decir, "If" ejecuta alg√∫n c√≥digo espec√≠fico despu√©s de comprobar si su expresi√≥n es verdadera.

```python
x = 0

if x < 0:
    x = 0
    print('Negative changed to zero')
elif x == 0:
    print('Zero')
elif x == 1:
    print('Single')
else:
    print('More')
```

```python
texto = 'Clase 00'

string_sabado = 'Clase del sabado'


if (len(texto) < len(string_sabado)) and (len(string_sabado) < 100):
    print('Ambas condiciones son verdaderas')
```

#### While
While se utiliza para la ejecuci√≥n repetida, siempre y cuando se cumpla alguna condici√≥n

```python
contador = 0
while contador < 3:
    print(contador)
    contador += 1  #contador = contador + 1
```

```python
dias_para_mi_cumple = 10
while dias_para_mi_cumple < 100:
    print('Faltan ' + str(dias_para_mi_cumple))
    print('dias para mi cumplea~os')
    dias_para_mi_cumple = dias_para_mi_cumple - 1
```

### Clases
- Los objetos son una encapsulaci√≥n de variables y funciones en una sola entidad.
- Los objetos toman sus variables y funciones de las clases. 
- Las clases son esencialmente una plantilla para crear sus objetos.

```python
class MiClase:

    def __init__(self):
        self.data = []
    
    def function(self):
        print("Este mensaje se imprime desde la clase")
```

```python
mi_objeto = MiClase()
mi_objeto.function()
```

```python
mi_objeto.data.append(123)
```

```python
mi_objeto.data
```
